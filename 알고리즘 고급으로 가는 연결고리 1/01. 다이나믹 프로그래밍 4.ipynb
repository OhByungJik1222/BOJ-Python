```python
import sys
sys.setrecursionlimit(10**6)

INF = float('inf')

def dfs(curr, visited):
    if visited == all:
        if graph[curr][start] != 0:
            return graph[curr][start]
        return INF
    
    if dp[curr][visited] != -1:
        return dp[curr][visited]

    tmp = INF
    for next in range(n):
        if graph[curr][next] != 0 and visited & (1 << next) == 0:
            tmp = min(tmp, graph[curr][next] + dfs(next, visited | (1 << next)))

    dp[curr][visited] = tmp

    return dp[curr][visited]

n = int(input())
graph = [list(map(int, input().split())) for _ in range(n)]

all = int('1' * n, 2)
dp = [[-1 for _ in range(1 << n)] for _ in range(n)]

start = 0
print(dfs(start, (1 << start)))
```

```python
import sys
sys.setrecursionlimit(10**6)

MOD = 1_000_000_000

def dfs(l, curr, visited):
    if l == n:
        if visited == all:
            return 1
        return 0
    
    if dp[l][curr][visited] != -1:
        return dp[l][curr][visited]
    
    tmp = 0
    if curr - 1 >= 0:
        tmp += dfs(l + 1, curr - 1, visited | (1 << (curr - 1)))
        tmp %= MOD
    if curr + 1 < 10:
        tmp += dfs(l + 1, curr + 1, visited | (1 << (curr + 1)))
        tmp %= MOD
    
    dp[l][curr][visited] = tmp

    return dp[l][curr][visited]

n = int(input())

all = int('1' * 10, 2)
dp = [[[-1 for _ in range((1 << 10))] for _ in range(10)] for _ in range(n + 1)]

ans = 0
for i in range(1, 10):
    ans += dfs(1, i, (1 << i))
    ans %= MOD
print(ans)
```

```python
import sys
# input = sys.stdin.readline

def check_x(x, bit):
    for i in range(m):
        if bit[i] == '1' and graph[x][i] == 'x':
            return False
    return True

def check_xy(curr_bit, prev_bit):
    for i in range(m):
        if curr_bit[i] == '1':
            for k in [-1, 1]:
                ni = i + k
                if 0 <= ni < m and prev_bit[ni] == '1':
                    return False
    return True

c = int(input())
for _ in range(c):
    n, m = map(int, input().split())
    graph = [list(input().strip()) for _ in range(n)]

    avail_bit = []
    for i in range(1 << m):
        bit = bin(i)[2:]
        bit = ('0' * (m - len(bit))) + bit

        cnt = 0
        can = 1
        for j in range(m):
            if bit[j] == '1':
                cnt += 1
                for k in [-1, 1]:
                    nj = j + k
                    if 0 <= nj < m:
                        if bit[nj] == '1':
                            can = 0
                            break
        if can:
            avail_bit.append([bit, cnt])

    dp = [[0 for _ in range(1 << m)] for _ in range(n)]
    for bit, cnt in avail_bit:
        if check_x(0, bit):
            dp[0][int(bit, 2)] = cnt
    
    for x in range(1, n):
        for curr_bit, curr_cnt in avail_bit:
            if check_x(x, curr_bit):
                for prev_bit, prev_cnt in avail_bit:
                    if dp[x - 1][int(prev_bit, 2)] and check_xy(curr_bit, prev_bit):
                        dp[x][int(curr_bit, 2)] = max(
                            dp[x][int(curr_bit, 2)],
                            curr_cnt + dp[x - 1][int(prev_bit, 2)]
                        )
    print(max(dp[-1]))
```

```python
import sys
sys.setrecursionlimit(10**6)

MOD = 9901

def dfs(col, row, curr, next):
    if col == m:
        dp[row + 1][next] += dp[row][curr]
        dp[row + 1][next] %= MOD
        return
    
    if curr & (1 << col):
        dfs(col + 1, row, curr, next)
    else:
        if col + 1 < m and not (curr & (1 << (col + 1))):
            dfs(col + 2, row, curr, next) # 가로
        dfs(col + 1, row, curr, next | (1 << col)) # 세로 (다음 행도 채워짐)

n, m = map(int, input().split())

dp = [[0 for _ in range(1 << m)] for _ in range(n + 1)]
dp[0][0] = 1

# bit[i] = 1 -> 이전 행에서 내려온 세로 도미노로 채워짐
# bit[i] = 0 -> 비어있으므로 가로또는 세로로 채워야 함 
for row in range(n):
    for curr in range(1 << m):
        if dp[row][curr]:
            dfs(0, row, curr, 0)
print(dp[n][0] % MOD)
```
