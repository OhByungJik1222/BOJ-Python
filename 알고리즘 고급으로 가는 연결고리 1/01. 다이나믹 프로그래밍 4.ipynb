14390번 타일 놓기 <span style="color:red">실패</span> - 2026.02.06

```python
INF = float('inf')

n, m = map(int, input().split())
graph = [list(input()) for _ in range(n)]

ALL = (1 << m) - 1

# dp[state]
# state: 길이 m비트
# 각 비트는 현재 칸의 왼쪽까지 처리했을 때
# 해당 열에 가로 타일이 이어지고 있는지 여부
# 1: 가로 타일이 이어짐
# 0: 이어지지 않음
curr_dp = [0 for _ in range(1 << m)]
for i in range(n):
    for j in range(m):
        prev_dp, curr_dp = curr_dp, [INF for _ in range(1 << m)]
        for k in range(1 << m): # 이전 상태에 대해 전이
            curr_col = (k << 1) & ALL
            curr_row = curr_col + 1

            # 현재 칸이 기둥인 경우
            if graph[i][j] == '#':
                curr_dp[curr_col] = min(curr_dp[curr_col], prev_dp[k])
                curr_dp[curr_row] = min(curr_dp[curr_row], prev_dp[k])
            # 현재 칸이 빈 칸인 경우
            else:
                # 위 칸과 세로로 연결되는지 확인
                # 위 칸이 존재하고, 위 칸이 빈칸이고, 이전 상태의 MSB가 0
                col_flag = (i > 0 and graph[i - 1][j] == '.' and k & 1 << (m - 1) == 0)
                # 왼쪽 칸과 가로로 연결되는지 확인
                # 왼쪽 칸이 존재하고, 왼쪽 칸이 빈칸이고, 이전 상태의 LSB가 1
                row_flag = (j > 0 and graph[i][j - 1] == '.' and k & 1 == 1)

                # 현재 칸을 세로로 덮기
                # 위에서 연결되고 있으면 개수 증가 x
                # 안되어 있다면 개수 증가 (새 타일 시작)
                curr_dp[curr_col] = min(
                    curr_dp[curr_col],
                    prev_dp[k] if col_flag else prev_dp[k] + 1
                )
                # 현재 칸을 가로로 덮기
                # 왼쪽에서 연결되고 있으면 개수 증가 x
                # 안되어 있다면 개수 증가 (새 타일 시작)
                curr_dp[curr_row] = min(
                    curr_dp[curr_row],
                    prev_dp[k] if row_flag else prev_dp[k] + 1
                )
print(min(curr_dp))
```

13976번 타일 채우기 2 <span style="color:red">실패</span> - 2026.02.07

```python
# 기존 타일 채우기 점화식: f(n) = 4f(n-2) - f(n-4)
# 이 식만 가지고 DP 진행 시 한칸씩 채우기 때문에 TLE
# 이를 행렬식으로 변환하면
# -> | f(n) | = |4 -1||f(n-2)| = |4f(n-2) - f(n-4)|
#    |f(n-2)|   |1  0||f(n-4)|   |     f(n-2)     |
#
# 이 행렬식으로 거듭제곱을 한다 -> 한 단계 앞으로 간다.
# ex) [f(2), f(0)] = [3, 1]
#     |4 -1| * |3| = |11| => [f(4), f(2)]
#     |1  0|   |1|   | 3| 
#     |4 -1| * |11| = |41| => [f(6), f(4)]
#     |1  0|   | 3|   |11|
#
# 왜 n//2 - 1만큼 제곱하는가?
# S(k) = [f(2k), f(2k-2)], k = n//2
# 초기 상태 => S(1) = [f(2), f(0)]
# 목표 상태 => S(k) = [f(n), f(n-2)]
# 따라서 S(1) -> S(k) => k - 1번 이동 => (n//2 - 1)번 이동
MOD = 1_000_000_007

def mat_pow(mat, exp):
    res = [[1, 0], [0, 1]]
    while exp > 0:
        if exp & 1:
            res = mul(res, mat)
        mat = mul(mat, mat)
        exp >>= 1
    return res

def mul(a, b):
    return [
        [
            (a[0][0] * b[0][0] + a[0][1] * b[1][0]) % MOD,
            (a[0][0] * b[0][1] + a[0][1] * b[1][1]) % MOD
        ],
        [
            (a[1][0] * b[0][0] + a[1][1] * b[1][0]) % MOD,
            (a[1][0] * b[0][1] + a[1][1] * b[1][1]) % MOD
        ]
    ]

n = int(input())

if n % 2 == 1:
    print(0)
else:
    f2, f0 = 3, 1
    mat = [[4, -1], [1, 0]]
    k = n//2

    mat = mat_pow(mat, k - 1)
    ans = (mat[0][0] * f2 + mat[0][1] * f0) % MOD
    print(ans)
```
