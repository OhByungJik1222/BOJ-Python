18287번 체스판 이동 <span style="color:red">실패</span> - 2026.02.18

```python
MOD = 1_000_000_007

def mat_pow(mat, exp):
    res = [[int(i == j) for j in range(m)] for i in range(m)]
    while exp:
        if exp & 1:
            res = mul(res, mat)
        mat = mul(mat, mat)
        exp >>= 1
    return res

def mul(a, b):
    res = [[0 for _ in range(m)] for _ in range(m)]
    for i in range(m):
        for k in range(m):
            if a[i][k] == 0:
                continue
            for j in range(m):
                res[i][j] = (res[i][j] + a[i][k] * b[k][j]) % MOD
    return res

n, m = map(int, input().split())

# A: 홀수 → 짝수
# B: 짝수 → 홀수
mat_a = [[0 for _ in range(m)] for _ in range(m)]
mat_b = [[0 for _ in range(m)] for _ in range(m)]
for i in range(m):
    for j in range(m):
        if abs(i - j) <= 1:
            mat_b[i][j] = 1
        if abs(i - j) <= 1 and (i % 2 != j % 2):
            mat_a[i][j] = 1

# 2행 이동 (홀 → 짝(A) 먼저)
mat = mul(mat_b, mat_a)

if n % 2 == 1:
    # f(n) = mat^((n-1)//2) * f(1)
    mat = mat_pow(mat, (n - 1)//2)
else:
    # f(n) = A * mat^((n-2)//2) * f(1)
    mat = mat_pow(mat, (n - 2)//2)
    mat = mul(mat_a, mat)

f1 = [1 for _ in range(m)]
fn = [0 for _ in range(m)]
for i in range(m):
    for j in range(m):
        fn[i] = (fn[i] + mat[i][j] * f1[j]) % MOD
print(sum(fn) % MOD)
```

18288번 팀 연습 <span style="color:red">실패</span> - 2026.02.25

```python
MOD = 1_000_000_007

n, k = map(int, input().split())

# dp[idx][a_cnt][last][c_used]
# idx   : idx번째 문제
# a_cnt : A 사용한 개수 mod k
# last  : 0 -> A, 1 -> B, 2 -> C
# c_used: C 사용 여부
can_a = k if k > 0 else 1
dp = [[[[0, 0] for _ in range(3)] for _ in range(can_a)] for _ in range(n + 1)]

if k > 0:
    dp[1][1 % k][0][0] = 1
dp[1][0][1][0] = 1
dp[1][0][2][1] = 1
for idx in range(1, n):
    for a_cnt in range(can_a):
        for last in range(3):
            for c_used in range(2):
                curr = dp[idx][a_cnt][last][c_used]
                if curr == 0:
                    continue

                if k > 0: # A
                    a_next = (a_cnt + 1) % k
                    dp[idx + 1][a_next][0][c_used] += curr
                    dp[idx + 1][a_next][0][c_used] %= MOD
                
                if last != 1: # B
                    dp[idx + 1][a_cnt][1][c_used] += curr
                    dp[idx + 1][a_cnt][1][c_used] %= MOD

                # C
                dp[idx + 1][a_cnt][2][1] += curr
                dp[idx + 1][a_cnt][2][1] %= MOD

ans = 0
for last in range(3):
    ans += dp[n][0][last][1]
    ans %= MOD
print(ans)
```
