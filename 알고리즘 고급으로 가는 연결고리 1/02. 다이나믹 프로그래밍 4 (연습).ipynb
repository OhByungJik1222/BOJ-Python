```python
INF = float('inf')

def dfs(visited):
    if dp[visited] is not None:
        return dp[visited]
    
    tmp = set()
    if (visited & (visited - 1)) == 0:
        idx = visited.bit_length() - 1
        tmp.add(arr[idx])
        dp[visited] = tmp
        return dp[visited]
    
    sub = (visited - 1) & visited
    while sub:
        sub1 = sub
        sub2 = visited ^ sub1

        set1 = dfs(sub1)
        set2 = dfs(sub2)

        for a in set1:
            for b in set2:
                tmp.add(a + b)
                tmp.add(a - b)
                tmp.add(a * b)
                if b != 0:
                    tmp.add(a//b)
        sub = (sub - 1) & visited

    dp[visited] = tmp

    return dp[visited]

def is_prime(val):
    if val < 2:
        return False
    
    i = 2
    while i * i <= val:
        if val % i == 0:
            return False
        i += 1
    return True

n = int(input())
arr = list(map(int, input().split()))

dp = [None for _ in range(1 << n)]

all = (1 << n) - 1
ans = dfs(all)

min_ans = INF
max_ans = -INF

for val in ans:
    if is_prime(val):
        min_ans = min(min_ans, val)
        max_ans = max(max_ans, val)

if min_ans == INF or max_ans == -INF:
    print(-1)
else:
    print(min_ans)
    print(max_ans)
```
