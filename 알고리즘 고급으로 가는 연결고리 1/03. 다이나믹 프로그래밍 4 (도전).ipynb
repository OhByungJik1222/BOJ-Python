```python
def dfs(h, x, last, arr):
    if x == h:
        cols.append(tuple(arr))
        return
    for num in range(last + 1, n + 1):
        arr.append(num)
        dfs(h, x + 1, num, arr)
        arr.pop()

while True:
    try:
        t = list(map(int, input().split()))
        k = t[0]
        l = t[1:]
    except:
        break

    n = int(input())

    height = []
    for col in range(l[0]):
        h = 0
        for row in range(k):
            if l[row] > col:
                h += 1
        height.append(h)

    column = []
    for h in height:
        cols = []
        dfs(h, 0, 0, [])
        column.append(cols)

    dp = {}
    for col in column[0]:
        dp[col] = 1
    
    for i in range(1, l[0]):
        new_dp = {}
        for prev, cnt in dp.items():
            for curr in column[i]:
                flag = True
                for j in range(len(curr)):
                    if curr[j] < prev[j]:
                        flag = False
                        break
                if flag:
                    new_dp[curr] = new_dp.get(curr, 0) + cnt
        dp = new_dp
    print(sum(dp.values()))
```
