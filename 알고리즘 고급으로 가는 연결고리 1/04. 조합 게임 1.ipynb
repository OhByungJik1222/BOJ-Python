```python
MAX = 5000

def mex(s):
    x = 0
    while x in s:
        x += 1
    return x

m = int(input())
k = int(input())
moves = list(map(int, input().split()))

g = [0 for _ in range(MAX + 1)]
for i in range(1, MAX + 1):
    s = set()
    for move in moves:
        if i - move >= 0:
            s.add(g[i - move])
    g[i] = mex(s)

for p in range(1, 1000):
    for i in range(3000, MAX):
        if g[i] != g[i - p]:
            break
    else:
        period = p
        break

start = 3000 - period

cy_cnt = 0
for i in range(1, min(m, start) + 1):
    if g[i] == 0:
        cy_cnt += 1

if m > start:
    cycle = (m - start)//period
    remain = (m - start) % period

    cycle_cnt = 0
    for i in range(start + 1, start + period + 1):
        if g[i] == 0:
            cycle_cnt += 1

    cy_cnt += cycle * cycle_cnt

    for i in range(start + 1, start + remain + 1):
        if g[i] == 0:
            cy_cnt += 1
print(cy_cnt)
```

```python
MAX = 100

def mex(s):
    x = 0
    while x in s:
        x += 1
    return x

g = [0 for _ in range(MAX + 1)]
for n in range(2, MAX + 1):
    s = set()
    for a in range(1, n):
        b = n - a
        s.add(g[a] ^ g[b])
    g[n] = mex(s)

n, m = map(int, input().split())
print('A' if g[n] ^ g[m] else 'B')
```
