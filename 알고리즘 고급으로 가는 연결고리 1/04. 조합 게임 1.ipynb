```python
n = int(input())

dp = [-1 for _ in range(n + 1)]
dp[1] = 0
if n >= 2: dp[2] = 1
if n >= 3: dp[3] = 0
for i in range(4, n + 1):
    if dp[i - 1] == 0 or dp[i - 3] == 0:
        dp[i] = 1
    else:
        dp[i] = 0
print('SK' if not dp[n] else 'CY')
```
