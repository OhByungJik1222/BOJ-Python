9659번 돌 게임 5 <span style="color:green">성공</span> - 2026.03.10

```python
MAX = 100

def mex(s):
    x = 0
    while x in s:
        x += 1
    return x

n = int(input())

moves = [1, 3]
g = [0 for _ in range(MAX + 1)]
for i in range(1, MAX + 1):
    s = set()
    for m in moves:
        if i - m >= 0:
            s.add(g[i - m])
    g[i] = mex(s)

for p in range(1, 30):
    for i in range(60, MAX):
        if g[i] != g[i - p]:
            break
    else:
        period = p
        break
print('SK' if g[n % period] else 'CY')
```
