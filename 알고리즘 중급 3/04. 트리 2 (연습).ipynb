15480번 LCA와 쿼리 <span style="color:red">실패</span> - 2025.12.08

```python
# 세 쌍의 LCA 중 가장 깊이가 깊은 노드를 선택하면
# 루트에서 가장 가까운 공통 조상을 구할 수 있다.
#       1
#      / \
#     2   3
#    / \   \
#   r   4   5
#          / \
#         u   v
# lca(r, u) = 1 (깊이 0)
# lca(u, v) = 5 (깊이 3) ✅
# lca(r, v) = 1 (깊이 0)
import sys
import math
sys.setrecursionlimit(10**5)
input = sys.stdin.readline

def dfs(curr):
    visited[curr] = 1
    for next in graph[curr]:
        if not visited[next]:
            visited[next] = 1
            parent[next][0] = curr
            depth[next] = depth[curr] + 1

            dfs(next)

def lca(a, b):
    if depth[a] > depth[b]:
        a, b = b, a

    diff = depth[b] - depth[a]
    for k in range(LOG + 1):
        if diff & (1 << k):
            b = parent[b][k]

    if a != b:
        for k in reversed(range(LOG + 1)):
            if parent[a][k] != parent[b][k]:
                a = parent[a][k]
                b = parent[b][k]

        return parent[a][0]
    else:
        return a
    

n = int(input())

LOG = int(math.log2(n))

graph = [[] for _ in range(n + 1)]
for _ in range(n - 1):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

parent = [[0 for _ in range(LOG + 1)] for _ in range(n + 1)]
depth = [0 for _ in range(n + 1)]
visited = [0 for _ in range(n + 1)]

dfs(1)

for k in range(1, LOG + 1):
    for x in range(1, n + 1):
        parent[x][k] = parent[parent[x][k - 1]][k - 1]

m = int(input())
for _ in range(m):
    r, u, v = map(int, input().split())

    ans = lca(r, u)
    tmp = lca(u, v)
    if depth[ans] < depth[tmp]:
        ans = tmp
    tmp = lca(r, v)
    if depth[ans] < depth[tmp]:
        ans = tmp

    print(ans)
```
