```python
# 수빈이의 방문여부를 time의 홀짝따라 따로 저장
# time = 3 위치 5, time = 4 위치 5는 완전히 다르기 때문
# 이후, 1 time 마다 수빈이를 먼저 이동시키고, visited에 저장
# time을 증가시킨 뒤 동생의 위치 계산하여 visited[time % 2][nk] = 1이면 만난 것
from collections import deque

LIMIT = 500_000

def bfs(n, k):
    q = deque()
    q.append([n, 0])
    
    visited = [[0 for _ in range(LIMIT + 1)] for _ in range(2)]
    visited[0][n] = 1
    
    time = 0
    while q:
        nk = k + time * (time + 1) // 2
        if nk > LIMIT:
            return -1
        
        if visited[time % 2][nk]:
            return time
        
        for _ in range(len(q)):
            n, t = q.popleft()
            nt = t + 1
            p = nt % 2

            for dn in [-1, 1, n]:
                nn = n + dn
                if 0 <= nn <= LIMIT and not visited[p][nn]:
                    visited[p][nn] = 1
                    q.append([nn, nt])
        time += 1
    return -1

n, k = map(int, input().split())

print(bfs(n, k))
```
