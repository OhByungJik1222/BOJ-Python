```python
import math

def segment_min(left, right, idx):
    if left == right:
        tree[idx] = num[left]
        return tree[idx]
    
    mid = (left + right) // 2
    tree[idx] = min(
        segment_min(left, mid, idx * 2), 
        segment_min(mid + 1, right, idx * 2 + 1)
    )
    return tree[idx]

def find_min(start, end, idx, left, right):
    if left > end or right < start:
        return float('inf')
    if left <= start and right >= end:
        return tree[idx]
    
    mid = (start + end) // 2
    return min(
        find_min(start, mid, idx * 2, left, right),
        find_min(mid + 1, end, idx * 2 + 1, left, right)
    )

n, m = map(int, input().split())
num = [int(input()) for _ in range(n)]

h = int(math.log(n, 2)) + 1
size = 2 ** (h + 1) - 1

tree = [0 for _ in range(size + 1)]
segment_min(0, n - 1, 1)

for _ in range(m):
    left, right = map(int, input().split())
    print(find_min(0, n - 1, 1, left - 1, right - 1))
```
