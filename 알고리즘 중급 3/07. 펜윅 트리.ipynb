```python
# 세그먼트 트리
def segment_sum(left, right, idx):
    if left == right:
        tree[idx] = arr[left]
        return tree[idx]
    
    mid = (left + right) // 2
    tree[idx] = segment_sum(left, mid, idx * 2) + segment_sum(mid + 1, right, idx * 2 + 1)
    return tree[idx]

def update_sum(start, end, node, idx, diff):
    if start > idx or idx > end:
        return tree[node]
    
    if start == end:
        tree[node] += diff
        return tree[node]
    
    mid = (start + end) // 2
    tree[node] = update_sum(start, mid, node * 2, idx, diff) + update_sum(mid + 1, end, node * 2 + 1, idx, diff)
    return tree[node]

def find_sum(start, end, idx, left, right):
    if left > end or right < start:
        return 0
    if left <= start and right >= end:
        return tree[idx]
    
    mid = (start + end) // 2
    return find_sum(start, mid, idx * 2, left, right) + find_sum(mid + 1, end, idx * 2 + 1, left, right)

n, m, k = map(int, input().split())
arr = [0 for _ in range(n + 1)]
tree = [0 for _ in range(4 * n)]

for i in range(n):
    arr[i + 1] = int(input())

segment_sum(1, n, 1)

for _ in range(m + k):
    a, b, c = map(int, input().split())

    if a == 1:
        diff = c - arr[b]
        arr[b] = c
        update_sum(1, n, 1, b, diff)

    elif a == 2:
        print(find_sum(1, n, 1, b, c))
```

```python
def update(x, y, diff):
    while x <= n:
        tmp = y
        while tmp <= n:
            tree[x][tmp] += diff
            tmp += tmp & -tmp
        x += x & -x

def prefix_sum(x, y):
    s = 0
    while x > 0:
        tmp = y
        while tmp > 0:
            s += tree[x][tmp]
            tmp -= tmp & -tmp
        x -= x & -x
    return s

def range_sum(x1, y1, x2, y2):
    return prefix_sum(x2, y2) - prefix_sum(x1 - 1, y2) - prefix_sum(x2, y1 - 1) + prefix_sum(x1 - 1, y1 - 1)

n, m = map(int, input().split())
arr = [[0 for _ in range(n + 1)]]
for _ in range(n):
    arr.append([0] + list(map(int, input().split())))

tree = [[0 for _ in range(n + 1)] for _ in range(n + 1)]
for x in range(1, n + 1):
    for y in range(1, n + 1):
        update(x, y, arr[x][y])

for _ in range(m):
    w = list(map(int, input().split()))

    if w[0] == 0:
        x, y, c = w[1:]
        diff = c - arr[x][y]
        arr[x][y] = c
        update(x, y, diff)

    elif w[0] == 1:
        x1, y1, x2, y2 = w[1:]
        print(range_sum(x1, y1, x2, y2))
```
