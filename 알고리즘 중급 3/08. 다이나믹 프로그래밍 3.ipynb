1328번 고층 빌딩 <span style="color:red">실패</span> - 2026.01.04

```python
MOD = 1_000_000_007

n, l, r = map(int, input().split())
dp = [[[0 for _ in range(n + 1)] for _ in range(n + 1)] for _ in range(n + 1)]
dp[1][1][1] = 1

# dp[i][j][k]를 구하는 법 
# (건물 i개가 있을 때 좌측에서 j개, 우측에서 k개 건물이 보이는 경우)
# dp[i - 1][j - 1][k]에서 건물 하나 추가, 좌측에서 하나 더 보이는 경우
# dp[i - 1][j][k - 1]에서 건물 하나 추가, 우측에서 하나 더 보이는 경우
# dp[i - 1][j][k]에서 건물 하나 추가, 그러나 양 끝이 아닌 사이에 추가되어
# 양쪽에서 보이는 건물의 개수에 차이가 없는 경우
# 위 경우 빌딩이 들어갈 수 있는 경우는 양 끝을 뺀 (i - 2)가지 경우가 있으므로 곱해준다.
for i in range(2, n + 1):
    for j in range(1, l + 1):
        for k in range(1, r + 1):
            dp[i][j][k] = (dp[i - 1][j - 1][k] + dp[i - 1][j][k - 1] + dp[i - 1][j][k] * (i - 2)) % MOD
print(dp[n][l][r])
```

3948번 홍준이의 친위대 <span style="color:red">실패</span> - 2025.01.05

```python
MAX = 20

# dp[i][j][k]
# i: 현재 수열의 길이
# j: 마지막 숫자가 i개 중에서 j번째로 작은 값
# k: 마지막 변화 방향 (0 -> 감소, 1 -> 증가)
# 즉, 길이가 i이고, 마지막이 j번째 값이며, 마지막 비교가 k인 지그재그 수열의 개수
dp = [[[0, 0] for _ in range(MAX + 1)] for _ in range(MAX + 1)]
dp[1][1][0] = 1
dp[1][1][1] = 1

for i in range(2, MAX + 1): # 길이 i-1짜리 수열 뒤에 숫자 하나를 붙이는 방식
    for j in range(1, i + 1): # 새로 붙일 숫자가 i개 중에서 j번째로 작은 값이라고 가정
        
        # 길이 i, 마지막이 j번째 값이고 마지막 비교가 증가인 수열의 개수
        # 이전 값이 k번째, 현재 값이 j번째 -> 따라서 k가 j보다 작아야함
        # -> range(1, j)
        # 현재 구하려는 수열이 증가하면 이전은 반드시 감소여야 한다
        # -> dp[i - 1][k][0]
        # 즉, 감소로 끝났던 모든 길이 i-1짜리 수열 중
        # 현재 값보다 작은 값으로 끝난 것들만 골라서
        # 현재값을 붙인다.
        for k in range(1, j):
            dp[i][j][1] += dp[i - 1][k][0]
        
        # 길이 i, 마지막이 j번째 값이고 마지막 비교가 감소인 수열의 개수
        # 이전 값이 j번째 이상, 최대는 i-1 -> 따라서 k가 j보다 커야함
        # -> range(j, i)
        # 이때 range(j + 1, i)가 아닌 이유
        # -> 이때 k, j는 실제값이 아닌 순위 (새로운 값이 들어오면 순위가 바뀜)
        # 현재 구하려는 수열이 증가하면 이전은 반드시 증가여야 한다
        # -> dp[i - 1][k][1]
        # 즉, 증가로 끝났던 모든 길이 i-1짜리 수열 중
        # 현재 값보다 큰 값으로 끝난 것들만 골라서
        # 현재값을 붙인다.
        for k in range(j, i):
            dp[i][j][0] += dp[i - 1][k][1]

ans = [0 for _ in range(MAX + 1)]
ans[1] = 1
for i in range(2, MAX + 1):
    for j in range(1, i + 1):
        ans[i] += dp[i][j][0] + dp[i][j][1]

t = int(input())
for _ in range(t):
    n = int(input())
    print(ans[n])
```

12911번 좋아하는 배열 <span style="color:red">실패</span> - 2025.01.06

```python
# 문제에서 주어진 조건대로 수열을 구성하는 것이 아닌
# 조건을 만족하지 않는 수열을 구하여 빼는 방식으로 풀이.
MOD = 1_000_000_007

n, k = map(int, input().split())

# dp[n][k]: 마지막 숫자가 k인 n자리 수열의 수 
dp = [[0 for _ in range(k + 1)] for _ in range(n + 1)]

# p_sum[i]: 길이가 i인 모든 유효한 수열 개수의 합 -> sum(dp[i][:k+1])
p_sum = [0 for _ in range(n + 1)]

# 숫자 i로 끝나는 1자리 수열은 1개
for i in range(1, k + 1):
    dp[1][i] = 1

# 모든 1자리 수열은 k개
p_sum[1] = k

for i in range(2, n + 1): # 길이 i
    for j in range(1, k + 1): # 마지막 숫자 j

        # 우선 길이가 i-1인 모든 수열 뒤에 j를 붙이면
        # 길이 i인 수열이 된다.
        dp[i][j] += p_sum[i - 1]

        # 그러나 길이가 i-1인 수열 중에 마지막 숫자가 j의 배수이면 안됨.
        for mul in range(j * 2, k + 1, j):
            dp[i][j] -= dp[i - 1][mul]
        
        # 마지막으로 길이 i인 전체 경우의 수 누적
        p_sum[i] += dp[i][j]
        p_sum[i] %= MOD
print(p_sum[n] % MOD)
```

13392번 방법을 출력하지 않는 숫자 맞추기 <span style="color:red">실패</span> - 2026.01.07

```python
import sys
sys.setrecursionlimit(10**6)

# idx: 현재 처리 중인 자리
# left: 이전까지 왼쪽으로 돌린 누적 횟수
def roll(idx, left):
    if idx == n:
        return 0
    
    # 계산된 상태라면 다시 계산 x
    if dp[idx][left] != -1:
        return dp[idx][left]
    
    dp[idx][left] = float('inf')

    # 현재 자리 숫자 계산
    # 이전까지 왼쪽으로 돌린 누적 횟수가 현재 자리 숫자에 영향을 주기때문
    if left:
        curr = (start[idx] + left) % 10
    else:
        curr = start[idx]

    # 왼쪽으로 돌리는 경우
    # 현재 숫자는 "curr", 목표는 "end[idx]"
    # 왼쪽 회전 필요 횟수: left_dp[curr][end[idx]]
    # 이 회전은 다음 자리에도 영향을 미치기 때문에 left에 더해준다.
    dp[idx][left] = min(
        dp[idx][left],
        roll(idx + 1, (left + left_dp[curr][end[idx]]) % 10) + left_dp[curr][end[idx]]
    )

    # 오른쪽으로 돌리는 경우
    # 현재 숫자는 "curr", 목표는 "end[idx]"
    # 오른쪽 회전 필요 횟수: right_dp[curr][end[idx]]
    # 이 회전은 다음 자리에 영향을 미치지 않기 때문에 left 유지
    dp[idx][left] = min(
        dp[idx][left],
        roll(idx + 1, left) + right_dp[curr][end[idx]]
    )

    # 위 두 연산을 비교하여 오른쪽, 왼쪽 회전 중 최소 비용 선택

    return dp[idx][left]

n = int(input())
start = list(map(int, input()))
end = list(map(int, input()))

dp = [[-1 for _ in range(10)] for _ in range(n)]

# left_dp[i][j]: i에서 j로 가기위한 최소 왼쪽 회전 횟수
# right_dp[i][j]: i에서 j로 가기위한 최소 오른쪽 회전 횟수
left_dp = [[-1 for _ in range(10)] for _ in range(10)]
right_dp = [[-1 for _ in range(10)] for _ in range(10)]
for i in range(10):
    for j in range(10):
        tmp = 0
        while True:
            if (i + tmp) % 10 == j:
                break
            tmp += 1
        left_dp[i][j] = tmp

        tmp = 0
        while True:
            if (i - tmp) % 10 == j:
                break
            tmp += 1
        right_dp[i][j] = tmp

print(roll(0, 0))
```
