4811번 알약 <span style="color:green">성공</span> - 2025.12.29

```python
while True:
    n = int(input())

    if n == 0:
        break

    dp = [[0 for _ in range(n + 1)] for _ in range(n + 1)]
    for i in range(1, n + 1):
        dp[i][0] = 1

    for i in range(1, n + 1):
        for j in range(1, i + 1):
            if dp[i][j] == 0:
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
    
    print(dp[n][n])
```

1915번 가장 큰 정사각형 <span style="color:red">실패</span> - 2026.01.01

```python
n, m = map(int, input().split())
graph = [list(input()) for _ in range(n)]

dp = [[0 for _ in range(m)] for _ in range(n)]

ans = 0
for i in range(n):
    for j in range(m):
        if graph[i][j] == '1':
            if i == 0 or j == 0:
                dp[i][j] = 1
            else:
                dp[i][j] = min(dp[i - 1][j], dp[i][j - 1], dp[i - 1][j - 1]) + 1
            ans = max(ans, dp[i][j])
print(ans**2)
```

```python
MAX = 1001
MOD = 1_000_000_009

dp = [[0 for _ in range(MAX)] for _ in range(MAX)]
dp[1][1:4] = [1, 1, 1]
dp[2][2:4] = [1, 2]
dp[3][3] = 1

for i in range(2, MAX): # 1,2,3 개수
    for j in range(4, MAX): # target
        dp[i][j] = sum(dp[i - 1][j - 3:j]) % MOD

t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    print(dp[m][n])
```

```python
MAX = 1001
MOD = 1_000_000_009

dp = [[0 for _ in range(MAX)] for _ in range(MAX)]
dp[1][1:4] = [1, 1, 1]
dp[2][2:4] = [1, 2]
dp[3][3] = 1

for i in range(2, MAX): # 1,2,3 개수
    for j in range(4, MAX): # target
        dp[i][j] = sum(dp[i - 1][j - 3:j]) % MOD

t = int(input())
for _ in range(t):
    n, m = map(int, input().split())

    ans = 0
    for i in range(1, m + 1):
        ans += dp[i][n]
    print(ans % MOD)
```
