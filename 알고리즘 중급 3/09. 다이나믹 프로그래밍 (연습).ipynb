2169번 로봇 조종하기 <span style="color:red">실패</span> - 2026.01.10

```python
INF = float('inf')

n, m = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]

dp = [[-INF for _ in range(m)] for _ in range(n)]
dp[0][0] = graph[0][0]

# 첫줄은 오른쪽으로만 가능
for y in range(1, m):
    dp[0][y] = dp[0][y - 1] + graph[0][y]

for x in range(1, n):
    left = [-INF for _ in range(m)]
    right = [-INF for _ in range(m)]

    # 왼 -> 오
    left[0] = dp[x - 1][0] + graph[x][0]
    for y in range(1, m):
        left[y] = max(left[y - 1], dp[x - 1][y]) + graph[x][y]

    # 오 -> 왼
    right[-1] = dp[x - 1][-1] + graph[x][-1]
    for y in range(m - 2, -1, -1):
        right[y] = max(right[y + 1], dp[x - 1][y]) + graph[x][y]

    for y in range(m):
        dp[x][y] = max(left[y], right[y])
print(dp[-1][-1])
```

```python
n, m, k = map(int, input().split())

graph = [[] for _ in range(n + 1)]
for _ in range(k):
    a, b, c = map(int, input().split())
    if a > b:
        continue
    graph[a].append([b, c])

dp = [[-1 for _ in range(m + 1)] for _ in range(n + 1)]
dp[1][1] = 0

for i in range(1, n + 1):
    for cnt in range(1, m + 1):
        if dp[i][cnt] < 0:
            continue

        for next, score in graph[i]:
            if cnt + 1 <= m:
                dp[next][cnt + 1] = max(dp[next][cnt + 1], dp[i][cnt] + score)
print(max(dp[n][1:]))
```

1983번 숫자 박스 <span style="color:red">실패</span> - 2026.01.15

```python
n = int(input())
board0 = list(map(int, input().split()))
board1 = list(map(int, input().split()))

num0 = [] # 윗줄의 숫자만
num1 = [] # 아랫줄의 숫자만
blank0 = 0 # 윗줄의 빈칸 개수
blank1 = 0 # 아랫줄의 빈칸 개수
for i in range(n):
    if board0[i] == 0:
        blank0 += 1
    else:
        num0.append(board0[i])

    if board1[i] == 0:
        blank1 += 1
    else:
        num1.append(board1[i])

if blank0 == n or blank1 == n:
    print(0)
else:
    # dp[i][a][b]: i번째 열까지 사용하고,
    #              윗줄에서 a개의 빈칸, 아랫줄에서 b개의 빈칸을 사용했을 때 
    #              얻을 수 있는 최대값
    # i: 열 번호
    # a: 윗줄에서 사용한 빈칸의 개수
    # b: 아랫줄에서 사용한 빈칸의 개수
    dp = [[[0 for _ in range(blank1 + 1)] for _ in range(blank0 + 1)] for _ in range(n)]
    dp[0][0][0] = num0[0] * num1[0]
    for i in range(1, n):
        for a in range(blank0 + 1):
            # i - a < 0: 빈칸 초과
            # i - a >= len(num0): 숫자 초과
            if i - a < 0 or i - a >= len(num0):
                continue

            for b in range(blank1 + 1):
                if i - b < 0 or i - b >= len(num1):
                    continue

                # i열에 위아래 둘다 숫자
                # 빈칸 개수 변화 x, 곱 추가
                dp[i][a][b] = dp[i - 1][a][b] + num0[i - a] * num1[i - b]
                # i열에서 위만 빈칸
                if a - 1 >= 0:
                    dp[i][a][b] = max(dp[i][a][b], dp[i - 1][a - 1][b])
                # i열에서 아래만 빈칸
                if b - 1 >= 0:
                    dp[i][a][b] = max(dp[i][a][b], dp[i - 1][a][b - 1])
    print(dp[-1][-1][-1])
```
