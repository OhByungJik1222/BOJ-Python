1994번 등차수열 <span style="color:red">실패</span> - 2026.01.19

```python
n = int(input())
nums = [int(input()) for _ in range(n)]
nums.sort()

# dp[i][diff]: i번째 원소로 끝나는 공차가 diff인 등차수열의 최대 길이
dp = [{} for _ in range(n)]

# i를 끝으로 하는 모든 부분수열을 만들기 위해
# 가능한 모든 이전 위치 j를 시도
ans = 1
for i in range(n):
    for j in range(i):
        diff = nums[i] - nums[j]

        # j에 이미 공차가 diff인 등차수열이 있다면
        # 끝에 i를 붙여서 길이를 1 늘릴 수 있다.
        if diff in dp[j]:
            dp[i][diff] = dp[j][diff] + 1
        # 없다면 2 -> 등차수열의 최소 길이는 2이기 때문이다.
        else:
            dp[i][diff] = 2
        
        ans = max(ans, dp[i][diff])
print(ans)
```

1947번 선물 전달 <span style="color:red">실패</span> - 2026.01.20

```python
# 완전 치환
MOD = 1_000_000_000

n = int(input())

# 1번 사람은 자기 선물 제외 (n - 1)개 가능 (k번 사람의 선물을 받았다.)
# k번 사람의 상황에 따라 달라짐.
# case 1: k번 사람이 1번 선물을 받았을 때
#         → (1 ↔ k), 남은 사람 수: n - 2, 남은 사람들은 또 자기선물 x
#         따라서 경우의 수 (n - 2)
# case 2: k번 사람이 1번 선물을 받지 않았을 때
#         → 1번 고정 k는 아직 못받음, 남은 사람 수: n - 1, 남은 사람들은 여전히 자기선물 x 
#         따라서 경우의 수 (n - 1)
# 1번 사람은 (n - 1)가지 선택가능, 그리고 각 선택마다 dp[n - 2] + dp[n - 1] 발생
# 따라서 점화식: dp[n] = (n - 1) * (dp[n - 2] + dp[n - 1])
dp = [0, 0, 1]
for i in range(3, n + 1):
    dp.append(((i - 1) * (dp[i - 2] + dp[i - 1])) % MOD)
print(dp[n])
```

2092번 집합의 개수 <span style="color:red">실패</span> - 2026.01.21

```python
MOD = 1_000_000

t, a, s, b = map(int, input().split())
nums = sorted(list(map(int, input().split())))

cnt = [0 for _ in range(t + 1)]
for num in nums:
    cnt[num] += 1

# dp[i][j]: 앞에서 i번째 숫자 종류까지 봤을 때
#           총 j개를 고르는 경우의 수
dp = [[0 for _ in range(b + 1)] for _ in range(t + 1)]
dp[0][0] = 1

for i in range(1, t + 1):
    for j in range(b + 1):
        for k in range(min(cnt[i], j) + 1):
            # i번째 숫자의 개수가 cnt[i]일 때
            # 0개부터 cnt[i]개까지 선택 가능
            dp[i][j] = (dp[i][j] + dp[i - 1][j - k]) % MOD

ans = 0
for k in range(s, b + 1):
    ans = (ans + dp[t][k]) % MOD
print(ans)
```

2172번 팰린드롬 경로 <span style="color:red">실패</span> - 2026.01.22

```python
n, l = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]

dx = [-1, -1, 0, 1, 1, 1, 0, -1]
dy = [0, 1, 1, 1, 0, -1, -1, -1]

# dp[k][x1][y1][x2][y2]: 길이가 k이고,
#                        시작점이 (x1, y1),
#                        끝점이 (x2, y2) 인 팰린드롬 경로 개수
dp = [[[[[0 for _ in range(n)] 
        for _ in range(n)]
        for _ in range(n)]
        for _ in range(n)]
        for _ in range(l+1)]

# 길이 1 초기값 (홀수 중심)
# 혼자 있는 한칸 -> 반드시 팰린드롬
for x in range(n):
    for y in range(n):
        dp[1][x][y][x][y] = 1

# 길이 2 초기값 (짝수 중심)
# 인접한 두 칸 -> 두 값이 같으면 팰린드롬
if l >= 2:
    for x in range(n):
        for y in range(n):
            for d in range(8):
                nx = x + dx[d]
                ny = y + dy[d]
                if 0 <= nx < n and 0 <= ny < n:
                    if graph[x][y] == graph[nx][ny]:
                        dp[2][x][y][nx][ny] += 1

# 양쪽으로 확장
# 경로를 직접 만들지 않고 양끝 좌표만 기억해서 팰린드롬 유지
for k in range(1, l - 1):
    if k + 2 > l:
        break

    for x1 in range(n):
        for y1 in range(n):
            for x2 in range(n):
                for y2 in range(n):
                    cnt = dp[k][x1][y1][x2][y2]
                    if cnt == 0:
                        continue

                    # 양쪽에서 동시 이동
                    # 왼쪽 끝과 오른쪽 끝을 각각 한칸 이동
                    for d1 in range(8):
                        nx1 =  x1 + dx[d1]
                        ny1 =  y1 + dy[d1]
                        if 0 <= nx1 < n and 0 <= ny1 < n:
                            for d2 in range(8):
                                nx2 = x2 + dx[d2]
                                ny2 = y2 + dy[d2]
                                if 0 <= nx2 < n and 0 <= ny2 < n:
                                    # 양쪽에서 추가된 숫자가 같다면
                                    if graph[nx1][ny1] == graph[nx2][ny2]:
                                        dp[k + 2][nx1][ny1][nx2][ny2] += cnt

# 길이 l인 모든 팰린드롬 경로
ans = 0
for x1 in range(n):
    for y1 in range(n):
        for x2 in range(n):
            for y2 in range(n):
                ans += dp[l][x1][y1][x2][y2]
print(ans)
```

12950번 팰린드롬 보행 <span style="color:green">성공</span> - 2026.01.23

```python
from collections import deque

INF = float('inf')

n, m = map(int, input().split())

graph = [[' ' for _ in range(n)] for _ in range(n)]
for _ in range(m):
    a, b, s = input().split()
    graph[int(a)][int(b)] = s
    graph[int(b)][int(a)] = s

q = deque()
dp = [[INF for _ in range(n)] for _ in range(n)]
for curr in range(n):
    for next in range(n):
        if graph[curr][next] != ' ':
            dp[curr][next] = min(dp[curr][next], 1)
            q.append([curr, next])

for curr in range(n):
    for next1 in range(n):
        if graph[curr][next1] != ' ':
            for prev in range(n):
                if graph[prev][curr] != ' ':
                    if graph[prev][curr] == graph[curr][next1]:
                        dp[prev][next1] = min(dp[prev][next1], 2)
                        q.append([prev, next1])
            
            for next2 in range(n):
                if graph[next1][next2] != ' ':
                    if graph[curr][next1] == graph[next1][next2]:
                        dp[curr][next2] = min(dp[curr][next2], 2)
                        q.append([curr, next2])

while q:
    curr, next1 = q.popleft()

    for prev in range(n):
        if graph[prev][curr] != ' ':
            for next2 in range(n):
                if graph[next1][next2] != ' ':
                    if graph[prev][curr] == graph[next1][next2]:
                        if dp[prev][next2] > dp[curr][next1] + 2:
                            dp[prev][next2] = dp[curr][next1] + 2
                            q.append([prev, next2])

print(dp[0][1] if dp[0][1] != INF else -1)
```
