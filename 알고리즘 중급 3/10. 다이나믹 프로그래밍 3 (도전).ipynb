1994번 등차수열 <span style="color:red">실패</span> - 2026.01.19

```python
n = int(input())
nums = [int(input()) for _ in range(n)]
nums.sort()

# dp[i][diff]: i번째 원소로 끝나는 공차가 diff인 등차수열의 최대 길이
dp = [{} for _ in range(n)]

# i를 끝으로 하는 모든 부분수열을 만들기 위해
# 가능한 모든 이전 위치 j를 시도
ans = 1
for i in range(n):
    for j in range(i):
        diff = nums[i] - nums[j]

        # j에 이미 공차가 diff인 등차수열이 있다면
        # 끝에 i를 붙여서 길이를 1 늘릴 수 있다.
        if diff in dp[j]:
            dp[i][diff] = dp[j][diff] + 1
        # 없다면 2 -> 등차수열의 최소 길이는 2이기 때문이다.
        else:
            dp[i][diff] = 2
        
        ans = max(ans, dp[i][diff])
print(ans)
```
